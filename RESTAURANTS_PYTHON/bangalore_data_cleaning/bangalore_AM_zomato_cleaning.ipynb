```python
import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
# importing data and creating copy of dataframe
df_oryginal = pd.read_csv('zomato.csv')
df = df_oryginal.copy()
df.info()
```

```python
# dropping unneccesary columns
df.drop(columns=['online_order', 'book_table', 'dish_liked', 'reviews_list', 'menu_item', 'listed_in(type)', 'listed_in(city)'], inplace=True)
df.info()
```

```python
# cenaming long column name
df.rename(columns={'approx_cost(for two people)': 'cost_for_2'}, inplace=True)
df.info()
```

```python
# cleaning restaurants names using regex pattern
names = df['name'].replace(['Ã.*©','.*©','Ã.*¨','Ã.*±','Ã.*¢','Ã.*°','Ã.*»','Ã.*ª','Ã.*'],['é','é','é','ñ','â','°','û','e','’'],regex=True)
df.update(names)
df.head(15)
```

```python
# cleaning rate values and change it to float type
df['rate'] = df['rate'].replace(['NEW','-'], [np.nan, np.nan])
df['rate'] = df['rate'].str.split('/').str[0]
df['rate'] = df['rate'].str.strip()
df['rate'] = df['rate'].astype(float)
df['rate'].unique()
```

```python
# cleaning number columns and split it into two columns
# in some casese two numbers provided
phone_numbers = df['phone'].str.split(pat="""\n""", n=2, expand=True)
phone_numbers.rename(columns={0:'phone1', 1:'phone2'}, inplace=True)

cols = ['phone1', 'phone2']
phone_numbers[cols].astype(str)

phone_numbers['phone1'] = phone_numbers['phone1'].str.replace('\W', '', regex=True)
phone_numbers['phone2'] = phone_numbers['phone2'].str.replace('\W', '', regex=True)
df['phone1'] = phone_numbers['phone1']
df['phone2'] = phone_numbers['phone2']
df = df.drop(columns=['phone'])
df[['phone1', 'phone2']].head()
```

```python
# changing 'cuisines' values to lower case
# for easier search for the future functionality
df['cuisines'] = df['cuisines'].str.lower()
```

```python
# setting geolocator - necessary for connecting with geopy server
geolocator = Nominatim(user_agent="app") 
```

```python
# changing one of the setting that causes a warning
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# creating a dataframe for finding geo cooridinates
unique_locations = df['location'].unique()
df_locations = pd.DataFrame(unique_locations)
df_locations = df_locations.rename(columns={0: 'location'})
df_locations['location'].replace('', np.nan, inplace=True)
df_locations.dropna(inplace=True)
df_locations['location_name'] = df_locations['location'] + ', Bengaluru, India'
```

```python
# finding geo coordinates and extracting latitude & longitude
df_locations['geopy_data'] = df_locations['location_name'].apply(geolocator.geocode)
df_locations['latitude'] = df_locations['geopy_data'].apply(lambda x: x.latitude if x != None else None)
df_locations['longitude'] = df_locations['geopy_data'].apply(lambda x: x.longitude if x != None else None)
df_locations = df_locations.drop(columns=['geopy_data'])
df_locations
```

```python
# geopy couldn't find coordinates for three locations
df_locations[df_locations.isna().any(axis=1)]
```

```python
# adding lacking coordinates manually
lacking_coordinates = pd.read_csv('coordinates_for_three_locations_not_found_by_geopy.csv', sep=";")
lacking_coordinates
```

```python
# mergin data and replacing nan values
df_locations_full = df_locations.merge(lacking_coordinates, on='location_name', how='left')
df_locations_full['latitude'] = df_locations_full['latitude'].fillna(value=df_locations_full['latitude_b'])
df_locations_full['longitude'] = df_locations_full['longitude'].fillna(value=df_locations_full['longitude_b'])
df_locations_full.info()
```

```python
# dropping unnecessary columns
df_locations_full = df_locations_full.drop(columns=['latitude_b','longitude_b'])
df_locations_full.info()
```

```python
# merging coordinates to base data frame
df = df.merge(df_locations_full, left_on='location', right_on='location',  how='left')
df = df.drop(columns=['location_name'])
df.info()
```

```python
# exporting clean dataframe to csv
df.to_csv('bangalore_AM_zomato_clean.csv', index=False)
```
